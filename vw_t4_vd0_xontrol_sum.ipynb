```python
def calculate_checksum(hex_data):
    # Преобразуем строку в список 16-ричных значений
    data = [int(hex_data[i:i+2], 16) for i in range(0, len(hex_data), 2)]

    # Суммируем все значения
    total_sum = sum(data)

    # Инвертируем сумму и оставляем только младшие 8 бит
    checksum = ~total_sum & 0xFF  # Получаем контрольную сумму (1 байт)

    return hex(checksum)

def split_into_blocks(hex_data):
    # Убираем пробелы
    hex_data = hex_data.replace(" ", "")

    # Разбиваем строку на блоки по два байта
    blocks = [hex_data[i:i+4] for i in range(0, len(hex_data), 4)]

    return blocks

def find_matching_checksum(chek_base_data, check_data_str):
    # Разбиваем данные для перебора на блоки по 4 байта

    check_data = split_into_blocks(check_data_str)
    base_data = ''.join(split_into_blocks(chek_base_data))


    for block in check_data:



        # Берем последние два байта из текущего блока
        last_two_bytes = block[-2:]


        # Вставляем последние два байта в начало base_data
        new_data = last_two_bytes + base_data


        # Расчитываем контрольную сумму для новых данных
        checksum = calculate_checksum(new_data)


        # Берем только последние два байта контрольной суммы
        checksum_last_byte = checksum[-2:]

        # Сравниваем контрольную сумму с правой частью данных (сравниваем с тем значением, которое было изначально)
        if block == checksum_last_byte+last_two_bytes:
            print(f"Найдено совпадение!Контрольная сумма: {checksum[2:]}, hex={block}")
            return new_data

    print("Совпадений не найдено.")
    return None


# Основные данные
base_data_str = '''4f 0d 1b 0d 18 00 3a 00 5a 02 c7 04 fe 05 6b 08
46 0d e1 0d 06 00 61 00 de 02 b8 05 23 07 fd 09
ad 0f 64 10
'''

# Данные для перебора (в формате hex)
# Данные для перебора (в строковом формате без пробелов)
data_str = '''01 00 0a c5 00 b8 6a 00 00 00 00 00 00 00 00 00
00 00 00 00 00 00 00 00 00 00 00 00 00 00 00 ff
ff ff ff ff ff ff ff ff ff ff ff ff ff ff ff ff
ff ff ff ff ff ff ff ff ff ff ff ff ff ff ff ff
ff ff ff ff ff ff ff ff ff ff ff ff ff ff ff ff
ff ff ff ff ff ff ff ff ff ff ff ff ff ff ff ff
00 00 00 00 00 00 00 ff 00 00 00 00 00 00 00 ff
00 00 00 00 00 00 00 ff 03 03 03 01 01 01 1e 14
00 80 03 00 00 00 00 00 14 ff ff ff ff ff ff ff
56 57 5a 37 5a 30 ff ff ff ff ff ff ff ff 56 57
5a 37 5a 30 ff ff ff ff ff ff ff ff 56 57 5a 37
5a 30 ff ff ff ff ff ff ff ff ff ff ff ff ff ff
58 58 58 58 58 58 58 58 58 58 58 58 58 58 58 58
58 ff 58 58 58 58 58 58 58 58 58 58 58 58 58 58
58 58 58 ff ff ff ff ff ff ff fc fa fc fa fc fa
fc fa fc fa fc fa fc fa fc fa 08 08 04 00 00 10
00 00 00 00 00 00 00 00 00 00 ff ff ff ff ff ff
ff ff ff ff ff ff ff ff ff ff ff ff 37 44 30 39
32 30 38 32 32 4a 20 20 54 34 56 30 31 39 00 00
c5 0a 36 4e 30 39 30 39 39 30 31 20 19 2a 00 00
22 02 01 01 09 64 80 01 dc 05 dc 05 1b 00 15 03
4f 0d 1b 0d c2 07 06 05 0a 0a 0a 0e 0a 10 00 00
4c 01 cd 00 8a 00 df 00 ff 0f 04 32 da 02 1e 14
c2 01 03 ff 01 04 00 00 00 00 00 00 05 50 08 0a
0f 32 ff ff ff ff 04 32 da 02 01 0f 6d 01 06 01
a8 fe 02 ff 00 00 01 05 03 05 05 03 01 ff ff ff
04 43 05 02 09 92 70 05 80 ff 30 30 31 35 39 35
42 5f 00 02 30 30 31 36 31 34 42 5f 00 02 30 30
31 35 39 38 5f 5f 00 01 00 00 f8 4b 4c 00 a6 02
c9 00 19 04 ff 80 03 00 40 11 4f 98 50 94 59 2f
f4 f7 1e ff fa 2f d1 f9 af ea f9 33 80 33 e0 01
0a 00 ff 00 05 41 05 06 40 27 00 66 50 60 50 60
3c 28 32 46 90 93 54 bd 47 f6 80 be 74 78 73 7a
00 7a 50 ae 25 4b 32 ff ff ff b4 1e 0f 04 07 00
18 00 3a 00 5a 02 c7 04 fe 05 6b 08 46 0d e1 0d
06 00 61 00 de 02 b8 05 23 07 fd 09 ad 0f 64 10
00 00 3c 03 95 0b 2d 13 11 00 92 02 77 09 ba 0f
9b 00 9b 00 9b 00 f1 00 3a 01 9f 01 f0 01 56 02
9d 00 9d 00 9d 00 f3 00 3b 01 9e 01 ef 01 56 02
70 02 70 02 70 02 e0 01 40 01 a0 00 50 00 20 00
20 00 50 00 a0 00 40 01 e0 01 70 02 70 02 70 02
19 00 8b 00 7e 01 e4 02 48 04 ad 05 ad 05 ad 05
85 00 ad 00 c0 00 e4 00 09 01 d3 01 70 02 94 02
e0 02 a6 03 78 04 10 04 e0 03 98 03 58 03 58 02
b8 01 90 01 40 01 28 00 f0 00 90 01 50 02 a0 03
e0 03 10 04 f0 00 90 01 50 02 60 03 e0 03 10 04
13 00 13 00 db 02 db 02 07 05 a1 05 ff dd bf 00
d4 30 80 3e 2c 4c d8 59 84 67 30 75 dc 82 88 90
34 9e e0 ab 0a 00 2a 00 46 00 60 00 78 00 8e 00
a4 00 b9 00 cd 00 e1 00 64 05 88 13 8c ff 90 90
90 90 90 90 90 94 94 64 90 90 90 88 88 90 90 88
90 90 90 90 98 90 90 88 90 90 90 70 8c 90 90 90
90 90 a0 90 a0 90 90 a0 90 90 90 90 90 90 70 88
88 88 90 90 8c 90 70 90 90 90 90 90 90 90 00 13
cf d3 d5 d2 d6 d6 d3 d1 d0 d3 c0 d1 cd cd cb d2
d2 d1 d1 00 00 00 00 00 00 00 00 00 00 00 00 00
00 00 00 00 00 00 00 00 00 00 00 00 00 00 00 00
00 00 00 00 19 00 ff ff 89 dc ff 01 a0 a4 22 33
00 ff 0f 0a 15 f6 00 01 56 28 95 52 d3 7b ff f4
ff f9 00 00 00 30 80 f0 80 f0 80 f0 80 f0 18 00
ff 80 ff bf ff ff 18 00 ff 80 ff bf ff ff 18 00
ff 80 ff bf ff ff 00 5d ff 61 ff 61 ff 61 ff 81
80 89 00 90 00 80 bf ff 02 35 9a ff 00 80 bf ff
02 35 9a ff 00 80 bf ff 03 3c 9d ff 00 80 bf ff
02 35 9a ff 9f 8f 8c 71 45 06 36 53 54 5f 61 66
87 69 68 5d 5b 56 c1 ff ff ff ff 00 50 0a 02 53
0a 00 00 04 04 04 04 ff ff 00 00 00 00 64 ff ff
ff ff ff ff ff ff 00 00 00 00 30 30 72 00 8f 00
ab 00 c8 00 e5 00 e5 00 2e 00 87 01 e1 02 3a 04
93 05 93 05 78 00 96 00 b4 00 d2 00 dc 00 f0 00
2e 00 87 01 e1 02 3a 04 ad 04 93 05 ff ff ff ff
ff ff ff ff 09 01 0b 0f 0f 01 01 01 2e 7f 01 0d
01 0b 01 09 07 ff 00 00 00 00 00 00 00 00 00 00
00 00 00 00 00 00 00 00 00 00 00 00 00 00 00 00
00 00 00 00 00 00 00 00 00 00 00 00 00 00 00 00
00 00 00 00 00 00 00 00 00 00 00 00 00 00 96 92
4a 00 20 01 ad f4 03 89 01 d0 10 a9 20 2c 14 09
f0 09 ad 4b 0a 8d 03 04 20 b0 d5 20 4f 29 f0 22
8e 0f 04 a9 01 8d 0f 1f 20 5c bd a9 02 8d 0f 1f
ad cd 08 8d 93 05 ad cb 08 8d 94 05 ad cc 08 8d
'''
# Склеиваем строки в одну
check_data_str = data_str.replace("\n", " ")

chek_base_data = base_data_str.replace("\n", " ")

# Пробуем найти совпадение
matching_data = find_matching_checksum(chek_base_data, check_data_str)
```
