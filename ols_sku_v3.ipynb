```python
# Шаг 1: Импорт необходимых библиотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns

# Устанавливаем стиль графиков
sns.set(style='darkgrid')
df = df_org.copy()
# Дополнительные данные, которые можно добавить на основе исходного кода
df['Package_Weight'] = df['Analytic1'].str.extract('(\d+)').astype(float)  # Извлечение веса упаковки из столбца Analytic1

brand_list = df['Brand'].unique().tolist()

# Присваиваем ASC значение каждому бренду
asc_values = {brand: idx for idx, brand in enumerate(brand_list)}
df['ASC'] = df['Brand'].map(asc_values)

# Шаг 3: Генерация необходимых переменных для модели

# Кодируем категориальные переменные (Cheese_Type и Brand)
df['Cheese_Type'] = df['Articul8']  # Используем 'Articul8' как тип сыра
df = pd.get_dummies(df, columns=['Cheese_Type', 'Brand'], drop_first=True)
# Преобразование bool переменных в int (0 или 1)
for col in df.columns:
    if df[col].dtype == bool:
        df[col] = df[col].astype(int)

# Определяем независимые переменные
X_columns = [
    'PRICE_PER_UNIT', 'SALES_VOL', 'SHARE_VOL_PROMO', 'Package_Weight', 'ASC'
] + [col for col in df.columns if col.startswith('Cheese_Type_') or col.startswith('Brand_')]

# Зависимая переменная
df['market_share'] = df.groupby(['Year', 'Month'])['SALES_VOL'].transform(lambda x: x / x.sum())

# Логарифм отношения долей рынка
df['base_market_share'] = df.groupby(['Year', 'Month'])['market_share'].transform('first')

df['log_share_ratio'] = np.log(df['market_share'] / df['base_market_share'])

# # Удаление строк с NaN значениями
# df.dropna(subset=X_columns + ['log_share_ratio'], inplace=True)

# # Проверка типов данных после преобразования
# print("Типы данных после преобразования:\n", df[X_columns].dtypes)


# Шаг 4: Подготовка данных для модели и оценка

# Разделение на зависимую и независимые переменные
Y = df['log_share_ratio']
X = df[X_columns]

# Добавляем константу
X = sm.add_constant(X)

# Оцениваем модель с помощью OLS (линейная регрессия)
model = sm.OLS(Y, X).fit()

# Выводим результаты модели
print("\nРезультаты оценки модели:")
print(model.summary())

# Сохраним коэффициенты модели
model_coefficients = model.params

# Шаг 5: Прогнозирование долей рынка на основе полученных коэффициентов

# Вычисляем теоретическую утилиту (utility) для каждого SKU
df['utility'] = model.predict(X)

# Экспонента утилиты
df['exp_utility'] = np.exp(df['utility'])

# Пересчет прогнозируемых долей рынка
df['pred_market_share'] = df.groupby(['Year', 'Month'])['exp_utility'].transform(lambda x: x / x.sum())

# Выводим средние прогнозируемые доли рынка по SKU
mean_market_shares = df.groupby('Sku_name')['pred_market_share'].mean().reset_index()
# print("\nСредние прогнозируемые доли рынка по SKU:")
# print(mean_market_shares.sort_values('pred_market_share', ascending=False))

# Шаг 6: Исключение SKU и пересчет (анализ чувствительности)

# Удаление определенного SKU (например, 'Сыр полутвёрдый BELFOR Флоу-пак 240 г 45,0%')
df_after = df[df['Sku_name'] != 'Сыр полутвёрдый Брест Литовск Российский Флоу-пак 200 г 50,0%'].copy()

# Пересчет прогнозируемых долей рынка
df_after['pred_market_share_after'] = df_after.groupby(['Year', 'Month'])['exp_utility'].transform(lambda x: x / x.sum())

# Шаг 7: Анализ и визуализация результатов
comparison = df.merge(df_after[['Year', 'Month', 'Sku_name', 'pred_market_share_after']], on=['Year', 'Month', 'Sku_name'], how='outer', suffixes=('', '_after'))
comparison['pred_market_share_after'] = comparison['pred_market_share_after'].fillna(0)

# Рассчитываем изменение долей рынка
comparison['market_share_change'] = comparison['pred_market_share_after'] - comparison['pred_market_share']

# # Визуализация изменения долей рынка
# plt.figure(figsize=(12, 6))
# for sku in df['Sku_name'].unique():
#     sku_data = comparison[comparison['Sku_name'] == sku]
#     plt.plot(sku_data['Month'], sku_data['market_share_change'], marker='o', label=sku)
# plt.xlabel('Месяц')
# plt.ylabel('Изменение прогнозируемой доли рынка')
# plt.title('Изменение прогнозируемой доли рынка по SKU')
# # plt.legend()
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Функция для построения горизонтальной диаграммы с фильтрацией меток по доле рынка
def plot_horizontal_bar_chart_with_filtered_labels(ax, sizes_before, sizes_after, labels, title, min_share=0.10):
    # Фильтрация меток, отображаем только те, у которых доля >= min_share в любом из наборов данных
    filtered_labels = [label if size_before >= min_share or size_after >= min_share else '' for label, size_before, size_after in zip(labels, sizes_before, sizes_after)]
    filtered_sizes_before = [size_before if size_before >= min_share else 0 for size_before in sizes_before]
    filtered_sizes_after = [size_after if size_after >= min_share else 0 for size_after in sizes_after]

    # Позиции по оси Y для меток
    y_pos = np.arange(len(filtered_sizes_before))

    # Построение горизонтальной столбчатой диаграммы для данных до исключения
    ax.barh(y_pos - 0.2, filtered_sizes_before, height=0.4, color='skyblue', label='До исключения SKU')
    # Построение горизонтальной столбчатой диаграммы для данных после исключения
    ax.barh(y_pos + 0.2, filtered_sizes_after, height=0.4, color='salmon', label='После исключения SKU')

    # Установка меток на оси Y
    ax.set_yticks(y_pos)
    ax.set_yticklabels(filtered_labels)
    ax.invert_yaxis()  # Инвертируем ось Y, чтобы самые большие значения были сверху
    ax.set_xlabel('Доля рынка')
    ax.set_title(title, fontsize=14)
    ax.legend()
    return ax

# Исключаемый SKU
sku_to_exclude = 'Сыр полутвёрдый Брест Литовск Российский Флоу-пак 200 г 50,0%'

# Пример данных, создайте ваш DataFrame (df) здесь
# df = pd.DataFrame(...)

# Пересчет долей рынка ДО исключения SKU
df['exp_utility'] = np.exp(df['utility'])
df['pred_market_share'] = df.groupby(['Year', 'Month'])['exp_utility'].transform(lambda x: x / x.sum())

# Сохранение долей рынка до исключения
market_share_before = df.groupby('Sku_name')['pred_market_share'].sum().reset_index()
market_share_before.rename(columns={'pred_market_share': 'pred_market_share_before'}, inplace=True)

# Исключение SKU из данных и пересчет долей рынка
df_after_exclusion = df[df['Sku_name'] != sku_to_exclude].copy()
df_after_exclusion['pred_market_share_after'] = df_after_exclusion.groupby(['Year', 'Month'])['exp_utility'].transform(lambda x: x / x.sum())

# Сохранение долей рынка после исключения
market_share_after = df_after_exclusion.groupby('Sku_name')['pred_market_share_after'].sum().reset_index()

# Сравнение до и после исключения SKU
comparison = pd.merge(market_share_before, market_share_after, on='Sku_name', how='outer')
comparison['pred_market_share_after'] = comparison['pred_market_share_after'].fillna(0)

# Рассчитываем суммарную долю рынка по SKU
total_market_share_before = comparison['pred_market_share_before'].sum()
total_market_share_after = comparison['pred_market_share_after'].sum()

# Нормализация долей рынка для диаграмм
comparison['pred_market_share_before'] = comparison['pred_market_share_before'] / total_market_share_before
comparison['pred_market_share_after'] = comparison['pred_market_share_after'] / total_market_share_after

# Извлечение данных для диаграмм
sizes_before = comparison['pred_market_share_before'].values
sizes_after = comparison['pred_market_share_after'].values
labels = comparison['Sku_name'].values

# Построение горизонтальных столбчатых диаграмм
fig, ax = plt.subplots(figsize=(12, 10))  # Увеличенный размер для удобного отображения меток

# Горизонтальная столбчатая диаграмма долей рынка до и после исключения SKU
plot_horizontal_bar_chart_with_filtered_labels(ax, sizes_before, sizes_after, labels, 'Сравнение долей рынка до и после исключения SKU', min_share=0.0)

# Общий заголовок
plt.suptitle(f'Сравнение долей рынка до и после исключения "{sku_to_exclude}"', fontsize=16)
plt.tight_layout()
plt.show()

# Вывод доли рынка для интересующего SKU
specific_sku_share = comparison[comparison['Sku_name'] == sku_to_exclude]
if not specific_sku_share.empty:
    print(f"\nДоля рынка '{sku_to_exclude}' до исключения: {specific_sku_share['pred_market_share_before'].values[0]:.4f}")
    print(f"Доля рынка '{sku_to_exclude}' после исключения: {specific_sku_share['pred_market_share_after'].values[0]:.4f}")
else:
    print(f"\nДоля рынка '{sku_to_exclude}' после исключения: 0 (SKU исключен)")
```

```python
# Функция для построения круговой диаграммы с фильтрацией меток по доле рынка
def plot_pie_chart_with_filtered_labels(ax, sizes, labels, title, min_share=0.10):
    # Фильтрация меток, отображаем только те, у которых доля >= min_share
    display_labels = [label if size >= min_share else '' for label, size in zip(labels, sizes)]

    # Построение круговой диаграммы
    wedges, texts, autotexts = ax.pie(sizes, labels=display_labels, autopct=lambda pct: f'{pct:.1f}%' if pct >= min_share * 100 else '',
                                      startangle=140, wedgeprops=dict(width=0.3), textprops=dict(size=10))

    # Установка заголовка диаграммы
    ax.set_title(title, fontsize=14)
    return ax

# Извлечение данных для построения диаграмм
sizes_org = mean_org['org_market_share'].values
sizes_pred = mean_market_shares['pred_market_share'].values
labels_org = mean_org.index.values
labels_pred = mean_market_shares['Sku_name'].values

# Построение круговых диаграмм с фильтрацией меток

fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# Круговая диаграмма на основе SHARE_VOL (new_df) с фильтрацией меток по доле
plot_pie_chart_with_filtered_labels(ax[0], sizes_org, labels_org, 'Доля рынка истина SHARE_VOL', min_share=0.10)

# Круговая диаграмма на основе предсказанной утилиты (market_share_before) с фильтрацией меток по доле
plot_pie_chart_with_filtered_labels(ax[1], sizes_pred, labels_pred, 'Доля рынка на основе Модельная(расчетная)', min_share=0.10)

# Общий заголовок
plt.suptitle('Сравнение долей рынка (Фильтрация меток < 10%)', fontsize=16)
plt.tight_layout()
plt.show()
```
